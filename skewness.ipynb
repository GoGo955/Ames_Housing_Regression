## SKEWNESS

Here we'll try log+1 transformation and boxcox +1 transformation to see, which one is better for the data. (we use +1 transformation because of many 0's in dataset)

```python
# select numerical features
numerical = AmesDfPrep.select_dtypes(include=np.number).columns
AmesNum = AmesDfPrep[numerical]
# highly skewed features
Skw = AmesNum.apply(lambda x: skew(x))
Skw = Skw[abs(Skw) > 0.5]
SkwFeats = Skw.index
```

```python
# At first let's check the skewnes of data
SkewDf = AmesDfPrep[SkwFeats].skew().reset_index().rename(columns={'index':'feature', 0:'skewness'})
```

```python
# Log+1 transform of the skewed numerical features
AmesNum[SkwFeats] = AmesNum[SkwFeats].apply(np.log1p,axis=1, inplace=True)

temp = AmesNum.skew().reset_index().rename(columns={'index':'feature', 0:'skew_log'})
SkewDf = pd.merge(SkewDf,temp)
```

```python
# Boxcox+1 transform of the skewed numerical features 
AmesDfPrepCox = AmesDfPrep.copy()
for feat in SkwFeats:
    AmesDfPrepCox[feat] = boxcox1p(AmesDfPrep[feat], boxcox_normmax(AmesDfPrep[feat] + 1))
    
CoxDf = AmesDfPrepCox.skew().reset_index().rename(columns={'index':'feature', 0:'skew_box'})
SkewDf = pd.merge(SkewDf,CoxDf)
```

```python
with pd.option_context('display.max_columns', None):
    display(SkewDf.transpose())
```

We can notice, that in some cases data transformations worsen the skewness (i.e. log+1 for LotFrontage), so we'll check that.

```python
SkewDf[abs(SkewDf['skew_log'])>abs(SkewDf['skewness'])]
```

With LotFrontage and BsmtUnfSF log+1 transformation actually did bad. 

```python
SkewDf[abs(SkewDf['skew_box'])>abs(SkewDf['skewness'])]
```

Boxcox+1 also worsen skewness in 2 cases, but the difference isn't that big, as it was with log+1.

```python
box = len(SkewDf[abs(SkewDf['skew_box'])<0.5])
log = len(SkewDf[abs(SkewDf['skew_log'])<0.5])
print(f'Numbers of features with significantly reduced skewness with log+1 transformation: {log}')
print(f'Numbers of features with significantly reduced skewness with box+1 transformation: {box}')
```
