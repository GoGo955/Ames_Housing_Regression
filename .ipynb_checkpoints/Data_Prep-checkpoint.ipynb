## SKEWNESS

Here we'll try log+1 transformation and boxcox +1 transformation to see, which one is better for the data.

```python
# select numerical features
numerical = AmesDfPrep.select_dtypes(include=np.number).columns
AmesNum = AmesDfPrep[numerical]
# highly skewed features
Skw = AmesNum.apply(lambda x: skew(x))
Skw = Skw[abs(Skw) > 0.5]
SkwFeats = Skw.index
```

```python
# At first let's check the skewnes of data
SkewDf = AmesDfPrep[SkwFeats].skew().reset_index().rename(columns={'index':'feature', 0:'skewness'})
```

```python
# Log transform of the skewed numerical features to lessen impact of outliers
AmesNum[SkwFeats] = AmesNum[SkwFeats].apply(np.log1p,axis=1, inplace=True)

temp = AmesNum.skew().reset_index().rename(columns={'index':'feature', 0:'skew_log'})
SkewDf = pd.merge(SkewDf,temp)
```

```python
AmesDfPrepCox = AmesDfPrep.copy()
for feat in SkwFeats:
    AmesDfPrepCox[feat] = boxcox1p(AmesDfPrep[feat], boxcox_normmax(AmesDfPrep[feat] + 1))
    
CoxDf = AmesDfPrepCox.skew().reset_index().rename(columns={'index':'feature', 0:'skew_box'})
SkewDf = pd.merge(SkewDf,CoxDf)
```

**Some new simple and basic features:**

```python
# Total Square Footage
AmesDfPrep['TotalSF'] = AmesDfPrep.TotalBsmtSF + AmesDfPrep.GrLivArea
AmesDfPrep['TotalFloorSF'] = AmesDfPrep['1stFlrSF'] + AmesDfPrep['2ndFlrSF']
AmesDfPrep['TotalPorchSF'] = AmesDfPrep.OpenPorchSF + AmesDfPrep.EnclosedPorch + \
                        AmesDfPrep['3SsnPorch'] + AmesDfPrep.ScreenPorch

# Total Bathrooms
AmesDfPrep['TotalBathrooms'] = AmesDfPrep.FullBath + .5 * AmesDfPrep.HalfBath + \
                          AmesDfPrep.BsmtFullBath + .5 * AmesDfPrep.BsmtHalfBath

# Booleans
AmesDfPrep['HasBasement'] = AmesDfPrep.TotalBsmtSF.apply(lambda x: 1 if x > 0 else 0)
AmesDfPrep['HasGarage'] = AmesDfPrep.GarageArea.apply(lambda x: 1 if x > 0 else 0)
AmesDfPrep['HasPorch'] = AmesDfPrep.TotalPorchSF.apply(lambda x: 1 if x > 0 else 0)
AmesDfPrep['HasPool'] = AmesDfPrep.PoolArea.apply(lambda x: 1 if x > 0 else 0)
AmesDfPrep['WasRemodeled'] = (AmesDfPrep['YearRemod/Add'] != AmesDfPrep.YearBuilt).astype(np.int64)
AmesDfPrep['IsNew'] = (AmesDfPrep.YearBuilt > 2000).astype(np.int64)
AmesDfPrep['WasCompleted'] = (AmesDfPrep.SaleCondition != 'Partial').astype(np.int64)
```

**We can encode some categorical features as ordered numbers when there order:**

```python
# Bsmnt feats
AmesDfPrep['BsmtCond'].replace({'None' : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5}, inplace=True)
AmesDfPrep['BsmtExposure'].replace({None : 0, "Mn" : 1, "Av": 2, "Gd" : 3}, inplace=True)
AmesDfPrep['BsmtFinType1'].replace({None : 0, "Unf" : 1, "LwQ": 2, "Rec" : 3, "BLQ" : 4,
                                    "ALQ" : 5, "GLQ" : 6}, inplace=True)
AmesDfPrep['BsmtFinType2'].replace({None : 0, "Unf" : 1, "LwQ": 2, "Rec" : 3, "BLQ" : 4,
                                    "ALQ" : 5, "GLQ" : 6}, inplace=True)
AmesDfPrep['BsmtQual'].replace({None : 0, "Po" : 1, "Fa" : 2, "TA": 3, "Gd" : 4, "Ex" : 5}, inplace=True)

# Grg 
AmesDfPrep['GarageCond'].replace({None : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5}, inplace=True)
AmesDfPrep['GarageQual'].replace({None : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5}, inplace=True)

# Other quality feats
AmesDfPrep['ExterCond'].replace({"Po" : 1, "Fa" : 2, "TA": 3, "Gd": 4, "Ex" : 5}, inplace=True)
AmesDfPrep['ExterQual'].replace({"Po" : 1, "Fa" : 2, "TA": 3, "Gd": 4, "Ex" : 5}, inplace=True)
AmesDfPrep['FireplaceQu'].replace({None : 0, "Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5}, inplace=True)
AmesDfPrep['HeatingQC'].replace({"Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5}, inplace=True)
AmesDfPrep['KitchenQual'].replace({"Po" : 1, "Fa" : 2, "TA" : 3, "Gd" : 4, "Ex" : 5}, inplace=True)

# The others 
AmesDfPrep['Functional'].replace({"Sal" : 1, "Sev" : 2, "Maj2" : 3, "Maj1" : 4, "Mod": 5, 
                                       "Min2" : 6, "Min1" : 7, "Typ" : 8}, inplace=True)
AmesDfPrep['LandSlope'].replace({"Sev" : 1, "Mod" : 2, "Gtl" : 3}, inplace=True)
AmesDfPrep['LotShape'].replace({"IR3" : 1, "IR2" : 2, "IR1" : 3, "Reg" : 4}, inplace=True)
AmesDfPrep['PavedDrive'].replace({"N" : 0, "P" : 1, "Y" : 2}, inplace=True)
AmesDfPrep['Street'].replace({"Grvl" : 1, "Pave" : 2}, inplace=True)
AmesDfPrep['Utilities'].replace({"ELO" : 1, "NoSeWa" : 2, "NoSewr" : 3, "AllPub" : 4}, inplace=True)
```

## We can also try to simplify some categorical features.
